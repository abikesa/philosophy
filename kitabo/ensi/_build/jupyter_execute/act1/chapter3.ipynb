````python
import networkx as nx
import matplotlib.pyplot as plt

# Creating a directed graph
G = nx.DiGraph()

# Adding nodes (categories from the diagram)
nodes = [
    "Economic & Environmental Inequity",
    "Racial & Ethnic Inequity",
    "Chronic Disability Inequity",
    "Access to Pediatric Rehabilitation",
    "Decreased availability of service providers",
    "Longer wait times to initiation of service",
    "Decreased referral despite higher injury severity",
    "Outcomes from Pediatric Rehabilitation",
    "Decreased motor and cognitive scores at discharge",
    "Lower quality of life",
    "Higher likelihood of functional deficit post injury",
    "Rehabilitation Service Utilization and Provision",
    "Decreased utilization of mental health services",
    "Decreased acute rehabilitation provision",
    "Increased utilization of educational services and tools post injury",
    "Decreased high quality care",
    "Decreased functional ability",
    "Increased morbidity"
]

# Adding edges (relationships)
edges = [
    ("Economic & Environmental Inequity", "Access to Pediatric Rehabilitation"),
    ("Racial & Ethnic Inequity", "Access to Pediatric Rehabilitation"),
    ("Chronic Disability Inequity", "Access to Pediatric Rehabilitation"),
    ("Access to Pediatric Rehabilitation", "Decreased availability of service providers"),
    ("Access to Pediatric Rehabilitation", "Longer wait times to initiation of service"),
    ("Access to Pediatric Rehabilitation", "Decreased referral despite higher injury severity"),
    ("Decreased availability of service providers", "Decreased high quality care"),
    ("Longer wait times to initiation of service", "Decreased functional ability"),
    ("Decreased referral despite higher injury severity", "Increased morbidity"),
    ("Access to Pediatric Rehabilitation", "Outcomes from Pediatric Rehabilitation"),
    ("Outcomes from Pediatric Rehabilitation", "Decreased motor and cognitive scores at discharge"),
    ("Outcomes from Pediatric Rehabilitation", "Lower quality of life"),
    ("Outcomes from Pediatric Rehabilitation", "Higher likelihood of functional deficit post injury"),
    ("Access to Pediatric Rehabilitation", "Rehabilitation Service Utilization and Provision"),
    ("Rehabilitation Service Utilization and Provision", "Decreased utilization of mental health services"),
    ("Rehabilitation Service Utilization and Provision", "Decreased acute rehabilitation provision"),
    ("Rehabilitation Service Utilization and Provision", "Increased utilization of educational services and tools post injury"),
    ("Decreased utilization of mental health services", "Decreased functional ability"),
    ("Decreased acute rehabilitation provision", "Decreased high quality care"),
    ("Increased utilization of educational services and tools post injury", "Increased morbidity")
]

# Adding the nodes and edges to the graph
G.add_nodes_from(nodes)
G.add_edges_from(edges)

# Plotting the graph
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color="skyblue", node_size=3000, edge_color="black", linewidths=1, font_size=10, font_weight="bold", arrows=True)
plt.title("Pediatric Rehabilitation Inequity Flow")
plt.show()
````
